```python
import pandas as pd
import numpy as np 
from apyori import apriori
coffee_df = pd.read_csv('coffee_data.csv', header = None)
coffee_df
```

```python
items = []
for i in range(0,23):
    items.append([str(coffee_df.values[i,j])for j in range(0,6)])
final_rule = apriori(items,min_support=0.2,min_confidence=0.6,
                     min_lift=1.2,min_length=2)
final_results = list(final_rule)
```

```python
final_results
```

```python
for item in final_results:
    pair = item[0]
    items = [x for x in pair]
    print('Rule:'+items[0]+'-->'+items[1])
    
    print('Support:'+str(item[1]))
    print('Confidence:'+str(item[2][0][2]))
    print('Lift:'+str(item[2][0][3]))
    
    print('===========================================')
    
```
