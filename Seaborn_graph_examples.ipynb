```python
# Data Visualisation using seaborn from sample data 


import pandas as pd 
import seaborn as sns
from matplotlib import pyplot as plt
sns.set(style = "ticks", color_codes = True)
data = pd.read_csv('countries.csv')
afg = data[data.country == 'Afghanistan']
alg = data[data.country == 'Algeria']
sns.pointplot(x= afg.year, y = afg.population, color = 'b')
sns.pointplot(x = alg.year, y = alg.population, color = 'r')
plt.grid(True)
plt.title('AFG VS ALG population')
```

```python
# Data Visualisation using seaborn from sample data bar graph 

data2 = pd.read_csv('5000 Sales Records.csv')
eu = data2[data2.Region == 'Europe']
asia = data2[data2.Region == 'Asia']
sns.barplot(x = eu.Item_Type, y = eu.Units_Sold,)

plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='small'  
)
plt.grid(True)
plt.title('Europe Data Showing Item Type and How much Units Sold')
```

```python
sns.barplot(x = asia.Item_Type, y = asia.Units_Sold)

plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='small'  
)
plt.grid(True)
plt.title('Asia Data showing item type and how much units sold')
```
