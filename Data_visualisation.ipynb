```python
Data = pd.read_csv('countries.csv')
```

```python
Data
```

```python
us= Data[Data.country == 'United States']
```

```python
us
```

```python
China = Data[Data.country == 'China']
```

```python
China
```

```python
plt.plot(us.year, us.population/ 10**6)
plt.plot(China.year, China.population/ 10**6)
plt.legend(['United States','China'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Usa Vs China population from 1950-2007')
plt.show()
```

```python
afg = Data[Data.country == 'Afghanistan']
```

```python
afg
```

```python
plt.plot(afg.year,afg.population/ 10**6,)
plt.plot(China.year,China.population/ 10**6)
plt.legend(['Afghanistan','China'])
plt.xlabel('year')
plt.ylabel('population')
plt.title('Afghanistan vs China Population from 1952-2007')
plt.show()
```
