```python
import pandas as pd
from matplotlib import pyplot as plt 
record_sales_1000 = pd.read_csv('1000 Sales Records.Csv')
```

```python
record_sales_1000
```

```python
se = record_sales_1000[record_sales_1000.Country == 'Senegal']
```

```python
se
```

```python
plt.scatter(se.Total_Revenue, se.Total_Cost)
plt.legend(['Senegal'])
plt.xlabel('Total Revenue')
plt.ylabel('Total Cost')
plt.show()
```

```python
import numpy as np
from numpy.polynomial.polynomial import polyfit 
```

```python
b,m = polyfit(se.Total_Revenue, se.Total_Cost, 1)
```

```python
plt.plot(se.Total_Revenue, se.Total_Cost,'.')
plt.plot(se.Total_Revenue, b + m * se.Total_Revenue, '-')
plt.legend(['Senegal'])
plt.xlabel('(Total Revenue)')
plt.ylabel('(Total Cost)')
plt.title('Graph showing Total Revenue Vs Total Cost In Senegal')
plt.show()
```
