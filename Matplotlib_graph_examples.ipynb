```python

#Creating stacked bar chart using Matplotlib on CSV files. 

import pandas as pd 

import numpy as np

from matplotlib import pyplot as plt 

sample_data = pd.read_csv('5000 Sales Records.csv')

asia = sample_data[sample_data.Region == 'Asia']

eu = sample_data[sample_data.Region == 'Europe']

width = 0.5 # the width of the bar 

ax1 = plt.subplot2grid((1,1), (0,0))
ax1.bar( asia.Item_Type, asia.Total_Profit,width, color = 'r')
ax1.bar(eu.Item_Type, eu.Total_Profit,width, color = 'b')
for label in ax1.xaxis.get_ticklabels(0):
    label.set_rotation(45)

plt.ylabel('Profit')

plt.xlabel('Item Type')

plt.legend(('Asia', 'Europe'))

plt.title('Asia Vs Europe profits from Items')
plt.grid(True)
plt.show()


```

```python
# create scatter plot from csv file using matplotlib 
country_data = pd.read_csv('countries.csv')

afg = country_data[country_data.country == 'Afghanistan']
alg = country_data[country_data.country == 'Algeria']

plt.plot(afg.year, afg.population , '--r')

plt.plot(alg.year, alg.population, 'o-g')

plt.xlabel('Year')

plt.ylabel('Population')

plt.legend(('Afg','Alg'))

plt.title('Afg vs Alg population over time')
plt.grid(True)
plt.show()
```

```python
# create horizontal bar chart using matplotlib 
plt.rcdefaults()
fig, ax = plt.subplots()

ax.barh(asia.Item_Type, asia.Total_Profit/10**6,width, color = 'r')
ax.barh(eu.Item_Type, eu.Total_Profit/10**6,width, color = 'b')

plt.ylabel('Item Type')
plt.xlabel('Total Profit(10**6)')
plt.legend(('Asia', 'Europe'))
plt.grid(True)
plt.title('Horizontal Bar')
plt.show()
```
