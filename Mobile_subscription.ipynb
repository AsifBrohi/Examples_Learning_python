```python
import pandas as pd 

import numpy as np 

from matplotlib import pyplot as plt 

mobile_subs = pd.read_csv('mobile_subs_fully_filtered.csv')

mobile_subs
```

```python
subs_postpaid = mobile_subs['Subscription_post_paid']

subs_pre_post = mobile_subs['Subscription_pre_and_post_paid']

subs_prepaid = mobile_subs['Subscription_pre_paid']

total_subs = mobile_subs ['Grand_Total']

year = mobile_subs['Year']
```

```python
plt.plot(year,subs_postpaid/10**6, 'or-')

plt.plot(year,subs_prepaid/10**6, 'vb-')

plt.plot(year,total_subs/10**6,'xg-')

plt.xlabel('Year')

plt.ylabel('Subscriptions (10**6)')

plt.legend(['Prepaid','Postpaid','Total_subs'])

plt.title('Changes Of Moblie Subscription from 2004-2017')

plt.show()
```
